```python
# /content/Asian-Hornet-2-1
# -> 변환 필요 0을 1로

def replace_in_file(file_path, old_str, new_str):
    fr = open(file_path, 'r')
    lines = fr.readlines()
    fr.close()

    fw = open(file_path, 'w')
    for line in lines:
        fw.write(line.replace(old_str, new_str, 1))
    fw.close()
```

```python
import shutil
import os

# 파일 옮기기
def move_file(from_path, to_path):
  filename_list, src_list, dst_list = [], [], []

  for file in os.listdir(from_path):
    filename_list.append(file)

  for file in filename_list:
    src = os.path.join(from_path, file)
    src_list.append(src)

  for file in filename_list:
    dst = os.path.join(to_path, file)
    dst_list.append(dst)

  for src, dst in zip(src_list, dst_list):
    shutil.move(src, dst)
```
