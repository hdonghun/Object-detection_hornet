```python
import shutil
import os

# 파일 옮기기
def move_file(from_path, to_path):
  filename_list, src_list, dst_list = [], [], []

  for file in os.listdir(from_path):
    filename_list.append(file)

  for file in filename_list:
    src = os.path.join(from_path, file)
    src_list.append(src)

  for file in filename_list:
    dst = os.path.join(to_path, file)
    dst_list.append(dst)

  for src, dst in zip(src_list, dst_list):
    shutil.move(src, dst)
```

```python
import shutil
import os

# 파일 옮기기
def move_file(from_path, to_path):
  filename_list, src_list, dst_list = [], [], []

  for file in os.listdir(from_path):
    filename_list.append(file)

  for file in filename_list:
    src = os.path.join(from_path, file)
    src_list.append(src)

  for file in filename_list:
    dst = os.path.join(to_path, file)
    dst_list.append(dst)

  for src, dst in zip(src_list, dst_list):
    shutil.move(src, dst)
```

```python
import onnx
import tensorflow as tf
from pathlib import Path

# ONNX 모델 로드
onnx_model_path = '/content/yolov5/runs/train/bee_yolov5s_results/weights/best.onnx'
onnx_model = onnx.load(onnx_model_path)

# ONNX 모델을 TensorFlow SavedModel로 변환
saved_model_path = '/content/yolov5/runs/train/bee_yolov5s_results/weights/saved_model'
tf_rep = tf2onnx.rewrite(onnx_model)
tf_rep.export_graph(saved_model_path, input_dict={tf_rep.graph.get_input_tensor_by_name('input').name: tf.constant([1, 640, 640, 3], dtype=tf.float32)})
```

학습된 모델을 저장하고 나중에 재사용하려면, 모델 파일을 보존하는 것이 필요합니다. 주어진 코드에서는 --name 옵션을 사용하여 학습된 모델의 결과물을 저장하는 디렉토리의 이름을 지정하고 있습니다. 이 디렉토리에는 학습 중에 생성된 가중치 파일과 다양한 로그 및 결과물이 저장됩니다.
